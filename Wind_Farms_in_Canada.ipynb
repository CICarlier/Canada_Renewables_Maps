```python
provinces = [Alberta, British_Columbia, New_Brunswick, Newfoundland_and_Labrador, Nova_Scotia, Ontario, Prince_Edward_Island, Quebec, Territories, Manitoba, Saskatchewan]
#canada_wind = reduce(lambda  left,right: pd.merge(left,right, how='outer'), provinces)
for province in provinces:
  province.Year = province.Year.astype("str")
  province['Capacity (MW)'] = province['Capacity (MW)'].astype("float")
  print(province.dtypes)
```

```python
canada_wind = reduce(lambda  left,right: pd.merge(left,right, how='outer'), provinces)
canada_wind
```

```python
locator = Nominatim(user_agent='osm', country_bias="Canada", timeout=30)
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
lat = []
long = []
for value in canada_wind.index:
  if canada_wind.Location[value][0].isdigit():
    lat.append(float(re.findall('(\d+\.\d+)°', canada_wind.Location[value])[0]))
    long.append(-float(re.findall('(\d+\.\d+)°', canada_wind.Location[value])[1]))
  else:
    location = locator.geocode(canada_wind.Location[value])
    lat.append(location[1][0])
    long.append(location[1][1])
lat
```

```python
canada_wind['Latitude'] = lat
canada_wind['Longitude'] = long
canada_wind
```

```python
import folium

m = folium.Map(location=[60, -90], zoom_start=5, tiles='Stamen Terrain') 

for plant in canada_wind.index:
    folium.Marker(location=[canada_wind["Latitude"].iloc[plant], canada_wind["Longitude"].iloc[plant]], popup=canada_wind.iloc[plant]['Name']).add_to(m)

m
```
